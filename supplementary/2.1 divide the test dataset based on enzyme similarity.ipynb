### generate the fasta files

```python
import pandas as pd
import numpy as np
ESP_train_df = pd.read_csv('ESP_train_df.csv', header = 0)
ESP_test_df = pd.read_csv('ESP_test_df.csv', header= 0)
print(ESP_train_df.shape, ESP_test_df.shape)
```

```python
total_df = pd.concat([ESP_train_df,ESP_test_df], axis=0)
fasta_file = "all_sequences.fasta"
with open(fasta_file, "w") as f:
    for i, seq in enumerate(total_df['Protein sequence']):
        sequence_id = f"sequence_{i+1}"  # 自动生成一个序列ID，例如 sequence_1, sequence_2, ...
        f.write(f">{sequence_id}\n")
        f.write(f"{seq}\n")
```

### extract the enzyme sequence based on CD-hit

```python
import os
import pandas as pd
os.getcwd()
# 读取包含蛋白序列的FASTA文件
fasta_file = "all_sequences_clustered_sequences_100.fasta"

# 初始化用于存储蛋白序列的列表
sequences = []

# 打开FASTA文件并读取内容
with open(fasta_file, "r") as file:
    lines = file.readlines()
    sequence = ""
    for line in lines:
        if line.startswith(">"):
            if sequence:
                sequences.append(sequence)
            sequence = ""
        else:
            sequence += line.strip()
    if sequence:
        sequences.append(sequence)
# 创建一个DataFrame来存储数据
data = pd.DataFrame({
    'sequence': sequences,
})
# 保存数据到Excel文件
data.to_excel("all_sequences_clustered_sequences_100.xlsx", index=False)
print(data.shape)
```
